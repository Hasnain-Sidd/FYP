```python
df4=pd.read_csv("Cleaned_Players_Stats&Position_Data.csv")
df4.head(10)
```

```python
df4.drop(['Unnamed: 0'],axis='columns',inplace=True)
```

```python
df4.head(10)
```

```python
plt.figure(figsize=(10,5))
comp_counts=df4.value_counts("Comp")
plt.bar(comp_counts.index,comp_counts.values,color='red',width=.5)
leagues_names=np.arange(0,5)
plt.xlabel("Leagues",fontsize=15)
plt.ylabel("Players In Leagues",fontsize=15)
#plt.xticks(leagues_names,('SERI-A',"English Premier League","Ligue-1","Bundes-Liga","La-Liga"),fontsize=10,color='black')
plt.title("Players played in different Leagues",fontsize=20,color="Black")
plt.grid(linestyle='solid',linewidth=0.6,color='black',axis='y')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(data=df4,x="90s",binwidth=1)
plt.show()
```

```python
df4['Age']=df4["Age"].astype('int')
df4["Born"]=df4['Born'].astype('int')
```

```python
df4.dtypes.to_dict()
```

```python
df4.select_dtypes("number").head()
```

```python
sns.boxplot(data=df4,x='Age',y='Comp', palette=['green', 'brown', 'red', 'purple', 'orange'],hue='Comp')
age_range=np.arange(17,41,3)
plt.xlabel("AGE",fontsize=15,color='black')
plt.ylabel("Club Leagues",fontsize=15)
plt.title("AGE RANGE",fontsize=20,color='black')
plt.xticks(age_range)
plt.show()
```

```python
comp_age=df4.groupby(['Comp'])['90s'].mean()
comp_age.plot(kind='bar',color='brown')
plt.xlabel("Club Leagues",fontsize=15,color='black')
plt.ylabel("Number of Players",fontsize=15,color='black')
plt.title("Players played in different Leagues",fontsize=15,color='black')
plt.grid(color='black',axis='y',linestyle='dotted',linewidth=1)
plt.xticks()
```

```python
df4.groupby('Comp').agg(
    playing_mean=('90s','mean'),
    playing_std=('90s','std')
)
```

```python
ax=sns.countplot(data=df4,x='position',palette=['Blue','purple','green'],hue='position',width=.5)

for bars in ax.containers:
    ax.bar_label(bars,color='black',fontsize=10)
plt.xlabel("Positons",fontsize=15,color='black')
plt.ylabel("Count Of Players",fontsize=15,color='black')
plt.title("Players Positions",fontsize=20,color='black')
plt.grid(axis='y',linestyle='solid',linewidth=.8,color='black')

plt.show()
```

```python
plt.figure(figsize=(20,7))
custom_colors = sns.color_palette('hsv', len(df4['sub_position'].unique()))
ax1=sns.countplot(data=df4,x='sub_position',width=.5,palette=custom_colors,hue='sub_position')

for bars in ax1.containers:
    ax1.bar_label(bars,color='black',fontsize=12)
plt.xlabel("Sub-Positons",fontsize=15,color='black')
plt.ylabel("Count Of Players",fontsize=15,color='black')
plt.title("Players Sub-Positions",fontsize=20,color='black')
plt.grid(axis='y',linestyle='solid',linewidth=.8,color='black')
plt.xticks(fontsize=10)
plt.show()
```
