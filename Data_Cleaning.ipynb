```python
checking_missing_values=len(df5) * 0.05
print(checking_missing_values)
```

```python
df5.isna().sum().to_dict()
```

```python
dropping_missong_values=df5.columns[(df5.isna().sum() <= checking_missing_values) & (df5.isna().sum() > 0)]
dropping_missong_values.to_list()
```

```python
df5.dropna(subset=dropping_missong_values,inplace=True)
```

```python
df5.isna().sum().to_dict()
```

```python
filling_missing_values=df5['G/SoT'].mean()
df5['G/SoT']=df5['G/SoT'].fillna(filling_missing_values)
```

```python
df5.isna().sum().to_dict()
```

```python
dup_cols=set()

for x in range(df5.shape[1]):
    base_col=df5.iloc[:,x]
    for y in range(x+1,df5.shape[1]):
        com_cols=df5.iloc[:,y]
        if base_col.equals(com_cols):
            dup_cols.add(df5.columns.values[y])

print(dup_cols)
```

```python
df5=df5.drop(labels=dup_cols,axis='columns')
```
